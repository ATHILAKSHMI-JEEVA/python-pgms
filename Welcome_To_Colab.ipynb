```
import matplotlib.pyplot as plt
categories=['A','B','C','D']
values=[10,15,7,12]

graph=plt.bar(categories,values)

graph[0].set_hatch('+')
graph[0].set_facecolor('skyblue')

graph[1].set_hatch('.')
graph[1].set_facecolor('red')

graph[2].set_hatch('*')
graph[2].set_facecolor('blue')

graph[3].set_hatch('/')
graph[3].set_facecolor('pink')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('bar chart with manual hatches and colors')

plt.show()
```
